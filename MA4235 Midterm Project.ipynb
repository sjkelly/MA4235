```julia
# Setup our functions 
function f(x)
    return (1-x[1])^2 + (x[2]-x[1]^2)^2 
end;

function g(x)
    return (1-x[1])^2 + abs(x[2]-x[1]^2) 
end;

function h(x)
    return (1/4)*abs(1-x[1]) + abs(x[2]-2*abs(x[1])+1) 
end;
```

The gradients are calculated as
* $\nabla f(x,y)=\left(
\begin{array}{c}
-2(1-x)-4x(y-x^2)\\
2(y-x^2)\\
\end{array}
\right)$ 

* $\nabla g(x,y)=\left(
\begin{array}{c}
2x+2x(\mathrm{sgn}(x^2-y))-2\\
-\mathrm{sgn}(x^2-y)\\
\end{array}
\right)$ 

* $\nabla h(x,y)=\left(
\begin{array}{c}
-2(1-x)+-4(y-x^2)\\
2(y-x^2)\\
\end{array}
\right)$ 

```julia
# Setup gradient functions
function grad_f(x)
    return [-2*(1-x[1])-4*x[1]*(x[2]-x[1]^2) ; 2*(x[2]-x[1]^2)]
end;

function grad_g(x)
    return [-2*x[1]+2*x[1]*sign(x[1]^2-x[2])-2 ; -sign(x[1]^2-x[2])]
end;

function grad_h(x)
    return [-2*(1-x[1])-4*x[1]*(x[2]-x[1]^2) ; 2*(x[2]-x[1]^2)]
end;
```

```julia
# Setup number of interations and initial search points
iterations = 10;
initial_point = [-0.5; -0.7];

# Setup interpolation values for graphing
n = 1000;
start = -2;
stop = 2;
x_space = linspace(start,stop,n);
y_space = linspace(start,stop,n);

# Create matrices of z values for contour plots
z_f = zeros(n,n);
z_g = zeros(n,n);
z_h = zeros(n,n);
for x = 1:n
    for y = 1:n 
        z_f[x,y] = f([x_space[x]; y_space[y]]);
        z_g[x,y] = g([x_space[x]; y_space[y]]);
        z_h[x,y] = h([x_space[x]; y_space[y]]);
    end
end
```

## Line Search
Find a critical point of $\phi_k(t)$, defined as 
$$\phi_k(t) = f(x^{(k)}-t \nabla(x^{(k)}))$$
$$\phi^{'}_{k}(t) = -\nabla f(x^{(k)} - t \nabla f(x^{(k)})) \cdot \nabla f(x^{(k)})$$

https://en.wikipedia.org/wiki/Root-finding_algorithm

## BFGS
https://en.wikipedia.org/wiki/BFGS_method

```julia
# Implement the BFGS algorithm and bracketing algoithm (line search)

function line_search(x, f::Function, grad_f::Function, t=1.0, mu=0, eta=1000, alpha=(1/10^4), beta=0.9)
    
    # Setup anonymous functions for finding t
    phi(t) = f(x-t*grad_f(x))
    phi_prime(t) = -dot(grad_f(x-t*grad_f(x)), grad_f(x))
    
    while true
        if phi(t) > (phi(0) + alpha*t*phi_prime(0)) # Armijo fails
            eta = t
            elseif phi_prime(t) < beta*phi_prime(0) # Weak Wolfe fails
            mu = t
        else 
            break # both armijo and weak wolfe hold
        end
        
        if eta > 0
            t = (mu + eta)/2
        else
            t = 2t
        end
    end
    
    return t
end;

function BFGS(x0, f::Function, grad_f::Function, C=eye(2), tolerence=0.01, max_iterations=50)
    k = 1;
    x = zeros(2, max_iterations);
    x[:,1] = x0;
    p = zeros(2, max_iterations);
    t = zeros(max_iterations);
    while norm(grad_f(x[:,k])) > tolerence
        p[:,k] = -C * grad_f(x[:,k])
        t[k] = line_search(x[:,k], f, grad_f)
        x[:,k+1] = x[:,k] .+ t[k]*p[:,k]
        if k >= max_iterations
            println("BFGS Algorithm stopped at max iterations")
            break
        end
        k += 1
    end
    println("BFGS Algorithm stopped at ", k, " iterations.");
    return x[:,1:k]
end;
```

```julia
# Run BFGS optimization method
x_f = BFGS(initial_point, f, grad_f)
#x_g = BFGS(initial_point, g, grad_g)
#x_h = BFGS(initial_point, h, grad_h)
```
