## Implementation notes
Created with iJulia.

```julia
# Setup our functions 
function f(x,y)
    return (1-x)^2+(y-x^2)^2 
end;

function g(x,y)
    return (1-x)^2+abs(y-x.^2) 
end;

function h(x,y)
    return (1/4)*abs(1-x)+abs(y-2*abs(x)+1) 
end;
```

The gradients are calculated as
* $\nabla f(x,y)=\left(
\begin{array}{c}
-2(1-x)-4x(y-x^2)\\
2(y-x^2)\\
\end{array}
\right)$ 

* $\nabla g(x,y)=\left(
\begin{array}{c}
2x+2x(\mathrm{sgn}(x^2-y))-2\\
-\mathrm{sgn}(x^2-y)\\
\end{array}
\right)$ 

* $\nabla h(x,y)=\left(
\begin{array}{c}
-2(1-x)+-4(y-x^2)\\
2(y-x^2)\\
\end{array}
\right)$ 

```julia
# Setup gradient functions
function grad_f(x,y)
    return x + y
end;

function grad_g(x,y)
    return x + y
end;

function grad_h(x,y)
    return x + y
end;
```

```julia
# Setup number of interations and initial search points
iterations = 10;
initial_point = [-0.5; -0.7];

# Setup interpolation values for graphing
n = 1000;
start = -2;
stop = 2;
x_space = linspace(start,stop,n);
y_space = linspace(start,stop,n);

# Create matrices of z values for contour plots
z_f = zeros(n,n);
z_g = zeros(n,n);
z_h = zeros(n,n);
for x = 1:n
    for y = 1:n 
        z_f[x,y] = f(x_space[x], y_space[y]);
        z_g[x,y] = g(x_space[x], y_space[y]);
        z_h[x,y] = h(x_space[x], y_space[y]);
    end
end

# Setup minimizers

x_f = zeros(2, iterations);
x_f[:,1] = initial_point;
x_g = x_f;
x_h = x_f;
```

```julia
# We are using pyplot for graphing
# https://github.com/stevengj/PyPlot.jl
using PyPlot;

# Set number of contour levels to visualize
contour_levels = 100;
```

```julia
contour(x_space,y_space,z_f,contour_levels); # Plot contours
plot(x_f[1,:]', x_f[2,:]',color="red", linewidth=1.0, linestyle="-"); # Plot optimization of x
title("Contour plot of f");
```

```julia
contour(x_space,y_space,z_g,contour_levels); # Plot contours
plot(x_g[1,:]', x_g[2,:]',color="red", linewidth=1.0, linestyle="-"); # Plot optimization of x
title("Contour plot of g");
```

```julia
contour(x_space,y_space,z_h,contour_levels); # Plot contours
plot(x_h[1,:]', x_h[2,:]',color="red", linewidth=1.0, linestyle="-"); # Plot optimization of x
title("Contour plot of h");
```
